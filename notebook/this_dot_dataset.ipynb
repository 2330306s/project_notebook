```python
def process_video(video_id, data_dir="datasets/this_dot_dataset", frame_context=1):
    print("Downloading video...")
    video = YouTube(f'https://www.youtube.com/watch?v={video_id}')
    title = video.title
    dest_path = os.path.join(data_dir,title)
    if dest_path is None:
        dest_path = os.getcwd()
    else:
        if not os.path.isdir(dest_path):
            os.makedirs(dest_path)
            print(f'Created the following directory: {dest_path}')
    
    video_path = video.streams.get_by_itag(136).download(dest_path)
    print("Downloading transcript...")
    transcript = YouTubeTranscriptApi.get_transcript("H81Tdrmz2LA")
    fe = FrameExtractor(video_path)
    frames = []
    for t in transcript:
        frame = fe.times2frames([t["start"]])[0]
        surrounding_frames = list(range(frame-frame_context, frame+frame_context+1))
        t["frame"] = frame
        t["surrounding_frames"] = surrounding_frames
        frames += surrounding_frames
    print("Extracting frames from transcript...")
    fe.extract_specific_frames(frames, img_name="frame", dest_path=os.path.join(dest_path,"frames"))
    print("performing OCR on frames...")
    for i, t in tqdm.tqdm(enumerate(transcript)):
        surrounding_text = []
        for frame in tqdm.tqdm(t["surrounding_frames"], leave=False):
            image = cv2.imread(f"{dest_path}/frames/frame_{frame}.jpg")
            try:
                text = pytesseract.image_to_string(image)
                surrounding_text.append(text)
            except Exception as e:
                print(f"Exception during OCR at frame {frame} for: {title}")
                print(e)
        t["surrounding_text"] = surrounding_text
        # save the json every 30 steps just in case
        if i % 30 == 0 or i == len(transcript)-1:
            with open(os.path.join(dest_path,"transcript_OCR.json"), 'w') as f:
                json.dump(transcript, f)
    return transcript
```

```python
def makeImageData(imgpath):
    img_req = None
    with open(imgpath, 'rb') as f:
        ctxt = b64encode(f.read()).decode()
        img_req = {
            'image': {
                'content': ctxt
            },
            'features': [{
                'type': 'DOCUMENT_TEXT_DETECTION',
                'maxResults': 1
            }]
        }
    return json.dumps({"requests": img_req}).encode()
```

```python
def requestOCR(url, api_key, imgpath):
  imgdata = makeImageData(imgpath)
  response = requests.post(ENDPOINT_URL, 
                           data = imgdata, 
                           params = {'key': api_key}, 
                           headers = {'Content-Type': 'application/json'})
  return response
```

```python
import numpy as np
from scipy.optimize import leastsq, minimize

def get_spot_grid(x_step, y_step, x, y, x_dim=40, y_dim=20):
    # [x,y,[x,y]]
    raster_grid= np.array(np.meshgrid(np.arange(x_dim)*x_step+x, np.arange(y_dim)*y_step+y)) \
            .transpose([1,2,0]).reshape((x_dim*y_dim,2))
    return raster_grid


def get_mean_distance(raster_grid, data_grid):
#     loss = np.mean(np.sqrt((grid_pred[:data_grid.shape[0]] - data_grid)**2))
    num_data = data_grid.shape[0]
    num_grid_data = raster_grid.shape[0]
    tiled_points = np.tile(data_grid,(num_grid_data,1,1)).reshape((num_grid_data,num_data,2)).transpose([1,0,2])
    tiled_grid = np.tile(raster_grid,(num_data,1,1))
#     print(tiled_points)
#     print(tiled_grid)
#     print(tiled_grid-tiled_points)
#     print(np.power(tiled_grid-tiled_points,2))
#     print(np.sum(np.power(tiled_grid-tiled_points,2),axis=-1))
#     print(np.sqrt(np.sum(np.power(tiled_grid-tiled_points,2),axis=-1)))
#     print(np.min(np.sqrt(np.sum(np.power(tiled_grid-tiled_points,2),axis=-1)),axis=1))
    loss = np.mean(np.min(np.sqrt(np.sum(np.power(tiled_grid-tiled_points,2),axis=-1)),axis=1))
    return loss


def err_func(params, data_grid):
    x_step, y_step, x, y = params
    raster_grid = get_spot_grid(x_step, y_step, x, y)
    loss = get_mean_distance(raster_grid, data_grid)
#     loss += np.abs(x_step - y_step*0.43)
    return loss
```
